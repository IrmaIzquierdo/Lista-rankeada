```python
data["DATA_RESUMEN"][0]
dataFin=[]
##dat_min=data["DATA_RESUMEN"].lower()
##print(len(data["DATA_RESUMEN"]))
for i in range(0,len(data["DATA_RESUMEN"])):
    dataFin.append(data["DATA_RESUMEN"][i].lower())
dataFin    
    
```

```python
from bertopic import BERTopic
from bertopic.vectorizers import ClassTfidfTransformer
import json
import pandas as pd
```

Instruccion para reducir palabras vacias

```python
ctfidf_model = ClassTfidfTransformer(reduce_frequent_words=True)
```

Ejecucion del modelo

```python
%%time
from bertopic import BERTopic
modelB = BERTopic(ctfidf_model=ctfidf_model,verbose=True,embedding_model='paraphrase-multilingual-MiniLM-L12-v2',
                  min_topic_size= 3)
headline_topics, _ = modelB.fit_transform(dataFin)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk 
nltk.download()
from nltk.corpus import stopwords
```

Actualizados el modelo eliminando las stopword en espaol

```python
vectorizer_model = CountVectorizer(stop_words=stopwords.words("spanish"))
modelB.update_topics(dataFin, vectorizer_model=vectorizer_model)
```

```python
Guardamos el modelo 
```

```python
modelB.save("model_final", serialization="safetensors", save_ctfidf=True)
```
