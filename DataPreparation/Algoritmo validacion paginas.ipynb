Codigo para la validacion de documentos. 


```python
import re
import os
import pandas as pd
ruta1='pdf_2020_2024/'
ruta2='finales/'
contenido=os.listdir(ruta1)
for i in range(len(contenido)):
    pdfFile=ruta1+contenido[i]
    print(pdfFile)
    reader=PdfReader(pdfFile,strict=False)
    numPages=len(reader.pages)
    if numPages>18:
        nom='A1000'+str(i)+'.pdf'
        shutil.copy(ruta1+contenido[i],ruta2+nom)   
```
